```python
import os
import  json
import numpy as np
from scipy.spatial import distance
```

```python
img_path_list = []
label_path_list = []

img_path  = '졸음운전 예방을 위한 운전자 상태 정보 영상\\Training\\image'
label_path = '졸음운전 예방을 위한 운전자 상태 정보 영상\\Training\\label'

label_dir_name_list = os.listdir(label_path)
label_dir_name_list.sort()
label_dir_name_list
```

```python
normal_state_label_dir_name_list = []

for label_dir_name in label_dir_name_list:
    if label_dir_name[11:13] == '01':
        normal_state_label_dir_name_list.append(label_dir_name)

normal_state_label_dir_name_list
```

```python
yawn_state_label_dir_name_list = []

for label_dir_name in label_dir_name_list:
    if label_dir_name[11:13] == '03':
        yawn_state_label_dir_name_list.append(label_dir_name)

yawn_state_label_dir_name_list
```

```python
drowsy_state_label_dir_name_list = []

for label_dir_name in label_dir_name_list:
    if label_dir_name[11:13] == '02':
        drowsy_state_label_dir_name_list.append(label_dir_name)

drowsy_state_label_dir_name_list
```

```python
left_eye_ratio_list = []
right_eye_ratio_list = []

left_eye_x_pos = []
left_eye_y_pos = []

right_eye_x_pos = []
right_eye_y_pos = []

for drowsy_state_label_dir_name in drowsy_state_label_dir_name_list:
    drowsy_state_label_file_name_list = os.listdir(label_path + os.sep + drowsy_state_label_dir_name)

    for normal_state_label_file_name in drowsy_state_label_file_name_list:
        with open(label_path + os.sep + drowsy_state_label_dir_name + os.sep + normal_state_label_file_name) as f:
            label_data = json.load(f)
            key_points = (label_data['ObjectInfo']['KeyPoints']['Points'])
            key_points = [float(i) for i in key_points]

            eye_x = np.array(key_points[36 * 2:48 * 2:2])
            eye_y = np.array(key_points[(36 * 2) + 1:(48 * 2) + 1:2])

            left_eye_center = key_points[68*2:(69*2)]
            right_eye_center = key_points[(69*2):(70*2)]

            left_eye_H_dist = distance.euclidean((eye_x[0], eye_y[0]), (eye_x[3], eye_y[3]))
            left_eye_V_dist = min(distance.euclidean((eye_x[1], eye_y[1]), (eye_x[5], eye_y[5])), distance.euclidean((eye_x[2], eye_y[2]), (eye_x[4], eye_y[4])))

            right_eye_H_dist = distance.euclidean((eye_x[6], eye_y[6]), (eye_x[9], eye_y[9]))
            right_eye_V_dist = min(distance.euclidean((eye_x[11], eye_y[11]), (eye_x[7], eye_y[7])), distance.euclidean((eye_x[10], eye_y[10]), (eye_x[8], eye_y[8])))

            left_eye_ratio = left_eye_V_dist / left_eye_H_dist
            right_eye_ratio = right_eye_V_dist / right_eye_H_dist

            left_eye_ratio_list.append(left_eye_ratio)
            right_eye_ratio_list.append(right_eye_ratio)

            left_eye_x_pos.append(left_eye_center[0])
            left_eye_y_pos.append(left_eye_center[1])

            right_eye_x_pos.append(right_eye_center[0])
            right_eye_y_pos.append(right_eye_center[1])

left_eye_ratio_list = np.array(left_eye_ratio_list)
right_eye_ratio_list = np.array(right_eye_ratio_list)

left_eye_x_pos = np.array(left_eye_x_pos)
left_eye_y_pos = np.array(left_eye_y_pos)

right_eye_x_pos = np.array(right_eye_x_pos)
right_eye_y_pos = np.array(right_eye_y_pos)

pass
```
